```python
import os
import pandas as pd
from sklearn.model_selection import train_test_split
from autogluon.vision import ImagePredictor
from sklearn.utils import resample
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix
from PIL import Image

print("Bibliotecas importadas correctamente.")
```

```python
# from google.colab import drive
# drive.mount('/content/drive')
```

```python
# Directorio principal de imgenes en Google Drive
images_dir = 'arcgis-survey-images'
```

```python
# Verificar si el directorio existe
if not os.path.isdir(images_dir):
    raise FileNotFoundError(f"El directorio '{images_dir}' no existe. Por favor, verifica la ruta.")
```

```python
# Obtener las clases a partir de los nombres de los subdirectorios
class_names = sorted([d for d in os.listdir(images_dir) if os.path.isdir(os.path.join(images_dir, d))])
print(f"Clases encontradas: {class_names}")

# Recopilar rutas de imgenes y etiquetas
data = []
for class_label in class_names:
    class_dir = os.path.join(images_dir, class_label)
    for img_name in os.listdir(class_dir):
        if img_name.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif')):
            img_path = os.path.join(class_dir, img_name)
            data.append({'image': img_path, 'label': class_label})
```

```python
# Crear un DataFrame
df = pd.DataFrame(data)
print(f"Total de imgenes: {len(df)}")
```

```python
# Dividir en conjuntos de entrenamiento y validacin (80% entrenamiento, 20% validacin)
train_df, valid_df = train_test_split(df, test_size=0.2, stratify=df['label'], random_state=123)
print(f"Entrenamiento: {len(train_df)} imgenes")
print(f"Validacin: {len(valid_df)} imgenes")
```

```python
# Manejo de clases desequilibradas en el conjunto de entrenamiento
class_counts = train_df['label'].value_counts()
print("Conteo de imgenes por clase en el conjunto de entrenamiento:")
print(class_counts)
```

```python
# Encontrar la clase con el mayor nmero de muestras
max_count = class_counts.max()
```

```python
# Aumentar las muestras de las clases minoritarias
dfs = []
for class_label, count in class_counts.items():
    df_class = train_df[train_df['label'] == class_label]
    if count < max_count:
        df_class = resample(df_class,
                            replace=True,     # Muestra con reemplazo
                            n_samples=max_count,  # Aumentar a max_count
                            random_state=123)
    dfs.append(df_class)
```

```python
train_df_balanced = pd.concat(dfs)
```

```python
# Mezclar el DataFrame balanceado
train_df_balanced = train_df_balanced.sample(frac=1, random_state=123).reset_index(drop=True)
print("Datos de entrenamiento despus del balanceo:")
print(train_df_balanced['label'].value_counts())
```

```python
# Definir el predictor de imgenes de AutoGluon
predictor = ImagePredictor()
```

```python
import os
import multiprocessing

# Configurar Jupyter para usar un solo hilo
os.environ['OPENBLAS_NUM_THREADS'] = '1'
os.environ['MKL_NUM_THREADS'] = '1'
os.environ['OMP_NUM_THREADS'] = '1'

# Configurar el mtodo de inicio de multiprocesamiento
multiprocessing.set_start_method('spawn', force=True)
```

```python
# Entrenar el modelo con configuraciones modificadas
print("Iniciando el entrenamiento con AutoGluon...")
predictor.fit(
    train_data=train_df_balanced,
    tuning_data=valid_df,
    hyperparameters={
        'epochs': 50,
        'num_workers': 0,  # Mantener num_workers en 0
        'batch_size': 32,  # Reducir el tamao del batch si es necesario
        'lr': 0.01,  # Ajustar la tasa de aprendizaje
        'optimizer': 'sgd',  # Cambiar el optimizador si es necesario
    },
    time_limit=7200,  # Limitar el tiempo de entrenamiento a 2 horas
    holdout_frac=0.1,  # Fraccin de datos para validacin
    hyperparameter_tune_kwargs={'num_trials': 10},  # Nmero de pruebas para ajuste de hiperparmetros
    nthreads_per_trial=4,  # Nmero de hilos por prueba
    ngpus_per_trial=0  # Nmero de GPUs por prueba (si est disponible)
)
```

```python
# Evaluar el modelo en el conjunto de validacin
print("Evaluando el modelo en el conjunto de validacin...")
test_metrics = predictor.evaluate(valid_df)
print("Mtricas de evaluacin:", test_metrics)
```

```python
# Obtener predicciones en el conjunto de validacin
print("Obteniendo predicciones en el conjunto de validacin...")
valid_predictions = predictor.predict(valid_df)
```

```python
# Aadir las predicciones al DataFrame de validacin
valid_df['pred_label'] = valid_predictions
```

```python
# Generar el reporte de clasificacin
print("Reporte de clasificacin:")
print(classification_report(valid_df['label'], valid_df['pred_label'], target_names=class_names))
```

```python
# Crear la matriz de confusin
conf_matrix = confusion_matrix(valid_df['label'], valid_df['pred_label'], labels=class_names)
plt.figure(figsize=(10, 8))
sns.heatmap(conf_matrix, annot=True, fmt='d', xticklabels=class_names, yticklabels=class_names, cmap='Blues')
plt.xlabel('Prediccin')
plt.ylabel('Etiqueta')
plt.title('Matriz de Confusin')
plt.show()
```

```python
# Mostrar algunas imgenes con sus predicciones
def show_predictions(predictor, df, num_images=5):
    samples = df.sample(n=num_images, random_state=42)
    for idx, row in samples.iterrows():
        img_path = row['image']
        true_label = row['label']
        image = Image.open(img_path).convert('RGB')

        pred_label = predictor.predict({'image': img_path})

        plt.figure(figsize=(4,4))
        plt.imshow(image)
        plt.title(f"Verdadero: {true_label}\nPrediccin: {pred_label.values[0]}")
        plt.axis('off')
        plt.show()
```

```python
# Mostrar 5 predicciones aleatorias
show_predictions(predictor, valid_df, num_images=5)
```

```python
# Guardar el modelo
predictor.save("autogluon_plague_classifier")
print("Modelo guardado en 'autogluon_plague_classifier'.")
```
